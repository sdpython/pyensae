# Display a diagram with blockdiag

[blockdiag](http://blockdiag.com/) is module which plots diagrams.

```python
from pyensae.graph_helper import draw_diagram
```

```python
img = draw_diagram("""
        blockdiag {
            A -> B -> C -> D;
            A -> E -> F -> G;
        }
        """)
img
```

The page [blockdiag - simple block-diagram image generator](http://blockdiag.com/en/blockdiag/index.html) explains the syntax used to describe a graph.

```python
draw_diagram("""
blockdiag admin {
   A -> B -> C -> D;
   A -> E;
   A -> H;

   // A and B belong to first group.
   group {
      A; B;
   }
}
""")
```
