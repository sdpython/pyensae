# Map with folium

[folium](http://folium.readthedocs.org/en/latest/index.html) is a nice module to display maps with [Open Street Map](http://openstreetmap.fr/).

```python
import folium
import json
```

```python
center = [48.862, 2.346]
paris = folium.Map(center, zoom_start=13)
folium.Marker(center, popup='Les Halles').add_to(paris)
paris
```
